```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(451)
```

```python
l = 100
mutation_rate = 1/l
x = np.random.randint(2, size = l)
iter = 1500
goal_sum = 100
```

```python
def flip_bit(bit):
  return bit == 0
```

## 4a.

```python
sum_end = []
x = np.random.randint(2, size = l)
count = 0
last_sum = 0
for i in range(iter):
  x_m = np.copy(x)
  for j in range(l):
    random = np.random.uniform()
    if random < mutation_rate:
      x_m[j] = flip_bit(x_m[j])
  if abs(goal_sum - np.sum(x_m)) < abs(goal_sum - np.sum(x)):
    x = x_m 
  sum_end.append(np.sum(x))
plt.plot(sum_end)
plt.xlabel("Iterations")
plt.ylabel("Fitness")
plt.show()
```
